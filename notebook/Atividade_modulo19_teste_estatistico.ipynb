### **Efetividade de Duas Estratégias de Ensino**

Uma instituição de ensino está conduzindo uma pesquisa comparativa entre duas estratégias de ensino de matemática . O estudo foi desenhado para avaliar objetivamente a eficácia pedagógica dessas abordagens.

#### **Desenho Experimental**

* **Estratégia A**: Método tradicional.(grupo controle)
* **Estratégia B**: Nova abordagem (grupo experimental)
Os dados observados são amostras de 50 avaliações de cada estratégia.

Para fins textuais, adotei nomes e classifiquei as mesmas para tornar a atividade mais contextualizada.

#### **Hipóteses:**

*   **Hipótese nula ($H_o$)**: A média das notas dos alunos na estratégia A é igual à média das notas dos alunos na estratégia B.
*   **Hipótese alternativa($H_1$)**: A média das notas na Estratégia B é maior do que a média das notas na Estratégia A.

#### **Análise Estatística**
Para comparação do desempenho:
* Teste Z para duas amostras independentes

* Nível de significância de 5% (α = 0.05)

* Hipótese alternativa unicaudal (μ_B > μ_A)

### **Objetivo:**
Determinar se há uma diferença significativa no desempenho médio dos alunos entre as duas estratégias.
#### **Objetivos Complementares**
1. Avaliar Teste de hipótese
2. Avaliar variancia e média das amostras
3. Calcular Teste Z e P-valor
4. Avaliar a Distribuição da estatística do teste Z


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.stats.weightstats import ztest
```

Os dados são criados a seguir:

```python
# Definindo médias e desvios padrão para as notas nas duas estratégias
mean_A = 70
std_A = 10

mean_B = 75
std_B = 12

# Gerando as amostras de notas para cada estratégia de ensino da nossa base
np.random.seed(0)  # Para garantir a reprodutibilidade dos resultados
sample_A = np.random.normal(loc=mean_A, scale=std_A, size=50)
sample_B = np.random.normal(loc=mean_B, scale=std_B, size=50)

#print("Notas da Estratégia A:", sample_A[:5])
#print("Notas da Estratégia B:", sample_B[:5])
```

### 1. Teste de Hipótese e Valor - P

A hipótese nula (**H0 :** $\mu_B = \mu_A$) do problema propõe a comparação das estratégias A e B sem qualificar (ou direcionar) qual delas é melhor ou pior. Porém, a hipótese alternativa (**H1 :** $\mu_B > \mu_A$)questiona se a média de B é maior que a de A. Dessa forma, o problema é unicaudal à direita. Neste caso, testamos se nossa variável de teste ( $\bar{x}$) por exemplo, é significativamente maior que a média hipotética($\mu_o$). O p-valor é então definido por:

$$P_{valor}= P(Z  \geqslant z)= 1-P(Z\leqslant z)$$

Observe que definimos ele novamente em função do caso 1. Então novamente precisamos calcular a probabilidade acumulada **P** até o valor encontrado em nosso teste z.

### 3. Teste Z e P Valor

A **Tabela 2** resumo os resultados dos testes estatisticos. Concluimos que **A estratégia de ensino B não é significativamente melhor que a A**, uma vez que o p-valor (0.063) é maior que o nível de significância adotado (0.063 > 0.05). Outra forma de analisar esse resultado é através do limiar de rejeição das hipóteses, dado pelo valor do Z crítico (1.645, para α = 0.05).
A estatística Z observada no teste foi de 1.527, valor que se mostra inferior ao limiar de significância (1.527 < 1.645). Portanto, **não se encontra na região de rejeição da hipótese nula (H₀)**. A Figura 1 ilustra graficamente essas posições e a relação entre os valores.

| Parâmetro                | Valor       | Comparação               | Decisão        | Interpretação                     |
|--------------------------|-------------|--------------------------|----------------|-----------------------------------|
| **Tipo de teste**        | Unicaudal à direita |                      |                | H₁: μ_B > μ_A                    |
| **Estatística Z**        | 1.527       |                          |                |                                   |
| **Z crítico (α=0.05)**   | 1.645       | 1.527 < 1.645            |                |                                   |
| **p-valor**             | 0.063       | 0.063 > 0.05             | Não rejeita H₀ | p ≥ α                             |
| **Conclusão**           |             |                          |                | Diferença não significativa       |
**Tabela 2**


```python
# 2.Cálculo das médias e variancias
m_a=np.mean(sample_A)
m_b=np.mean(sample_B)
var_a=np.var(sample_A,ddof=1)
var_b=np.var(sample_B,ddof=1)

# Resultados
print("="*50)
print(f"\n{'Estatística':<15} | {'Estratégia A':>12} | {'Estratégia B':>12}")
print("-"*50)
print(f"{'Média':<15} | {m_a:>12.3f} | {m_b:>12.3f}")
print(f"{'Variância':<15} | {var_a:>12.3f} | {var_b:>12.3f}")


# 3) Teste Z e Valor P
# Variancia combinada
#n=len(sample_A)
#VC = np.sqrt(var_a/n + var_b/n) 
# Teste Z
#z = (m_b - m_a) / VC
#p_valor= 1-stats.norm.cdf(abs(z)) 

# Caso unilateral à direita (  H₁: μ_B > μ_A )
z, p_valor = ztest(sample_B, sample_A, alternative='larger') 
alpha=0.05
z_critico = stats.norm.ppf(1-alpha)
# Resultados
print("="*60)
print("\nTeste Z Unicaudal à Direita: H₁: μ_B > μ_A")
print("-"*60)
print(f"{'Estatística Z:':<20} {z:.3f}")
print(f"{'p-valor:':<20} {p_valor:.3f}")
print(f"{'z critico:':<20} {z_critico:.3f}")
print("-"*60)
# Condição
if z < alpha:
    print("Conclusão:  Rejeita H₀ (p < 0.05)")
    print("          → Estratégia B tem desempenho significativamente maior")
else:
    print("Conclusão:  Não rejeita H₀ (p ≥ 0.05)")
    print("          → Não há diferencas significativas nas estratégias")
print("="*60)
```

### 4. Distribuição da estatística do teste Z
A linha azul tracejada, que representa o valor observado de Z (**~1.542**), posiciona-se próxima ao valor crítico (**1.645**), mas sem ultrapassá-lo. Isso indica que a diferença de desempenho entre as estratégias A e B está no limiar da significância estatística, mas não atinge o nível necessário (α = 0.05) para rejeitar a hipótese nula (H₀). Z observado está na cauda direita da distribuição indicando que a Estratégia B apresentou um desempenho numericamente superior,mas não suficientemente para um nivel de confiança de 95%. Uma resultado que reforça isso, é o p-valor( 0.061) ligeiramente acima da condição α = 0.05. reforçando que a diferença não é estatisticamente significativa.



```python
# 4 Distribuição da estatística do teste Z

fig, ax = plt.subplots(figsize=(10, 5))

# Parâmetros da distribuição normal padrão (Z)
mu = 0       
sigma = 1   

# Valores de X e Y para a distribuição
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
y = stats.norm.pdf(x)  # PDF da normal padrão (Z)

# Teste Z e p valor
z_statistic, p_value = ztest(sample_B, sample_A, alternative='larger') 

# valor crítico para Z (unicaudal à direita, alpha=0.05)
z_critico = stats.norm.ppf(0.95)  # Percentil 95% da normal padrão

# Gráfico da densidade
ax.plot(x, y, 'k',label='Distribuição Normal Padrão (Z)')

# Linha do valor observado
ax.vlines(x=[z_statistic], ymin=0, ymax=y.max(), 
          colors='dodgerblue', linestyles='dotted', label=f'Z observado = {z_statistic:.2f}')

ax.vlines(x=[z_critical], ymin=0, ymax=y.max(), 
          colors='tomato', linestyles='--', label=f'Z Critico = {z_critico:.2f}')

# Área crítica (unicaudal à direita)
plt.fill_between(x, y, where=(x > z_critical), 
                 alpha=0.9, color='tomato', label='Área crítica (α=0.05)') 
ax.set_ylim(0, max(y)*1.05)

# Configurações do gráfico
plt.legend()
plt.xlabel('Valores de Z')
plt.ylabel('Densidade de Probabilidade')
plt.title('figura 1: Teste Z Unicaudal à Direita: Região Crítica e Valor Observado')
plt.grid(True)
plt.show()
```
