## Objetivos: Identificar váriaveis que impulsionem o limite de crédito. 

### Modificação no conjunto de dados original
1. Os valores das colunas Salário e Limite_Crédito, originalmente representados como strings numéricas, foram convertidos para um formato numérico apropriado
2. Foram criadas novas variáveis categóricas a partir das colunas originais. Para a coluna Histórico de Inadimplência, foram geradas as variáveis Bom Pagador e Inadimplência. A coluna Imóvel Próprio foi transformada nas variáveis Com Imóvel e Sem Imóvel. Por fim, a variável Estado Civil foi representada pelas categorias Casado e Solteiro. 

### Pré-avaliação

De posse disso, A **figura 1**, apresenta  relação das correlações com as seguintes variáveis: Idade, Salário, Sem Imóvel, Com Imóvel, Bom Pagador, Inadimplência, Casado, Solteiro e Limite de Crédito. Essa análise possibilitou a identificação de padrões e relações entre as variáveis. 

* **Correlação positiva com limite de crédito** : Idade, Salário, ter imovel,Bom Pagador, Casado
* **Correlação Inversa  com limite de crédito** : Sem Imovel,Inadiplencia, Solteiro

De fato, observamos valores opostos nas correlações entre as variáveis relacionadas ao estado civil, histórico de inadimplência e posse de imóvel. Tal resultado era esperado uma vez que indivíduos casados tendem a apresentar maior estabilidade financeira e comprometimento com obrigações de crédito, refletindo-se em um maior limite de crédito concedido. Da mesma forma, a posse de imóvel e um histórico de bom pagador são indicativos de solidez financeira, o que justifica a correlação positiva com o limite de crédito. Em contraste, indivíduos solteiros e com histórico de inadimplência são frequentemente percebidos como perfis de maior risco, resultando em um menor limite de crédito. Respondemos duas perguntas.

#### 1) Qual a diferença de crédito entre os grupos?

A concessão de limite de crédito varia entre diferentes grupos.Constatamos que indivíduos casados, quando comparado com solteiros, apresentam um limite 254% maior (**figura 2**). De forma semelhante, o limite de crédito para bons pagadores, quando comparados com  indivíduos com algum histórico de inadimplência,  apresentam limite 284% maior.

* **Solteiros X Casados** : Pessoas casadas tem um limite de crédito médio 254% maior que pessoas solteiras
* **Bom pagador x Inadimplentes  com limite de crédito** : Bons pagadores tem um limite de crédito 284% maior que pessoas com histórico de indadimplência 


#### 2) Qual a relação entre salário e Limite de Crédito?

O limite de crédito apresenta uma relação crescente com o salário. No entanto, variáveis como inadimplência e estado civil (ser solteiro) impactam negativamente a concessão de um limite mais alto. Observa-se (**figura 3**) que, mesmo indivíduos solteiros com salários elevados tendem a possuir um limite de crédito inferior quando comparados a pessoas casadas ou classificadas como bons pagadores.

### Considerações Finais
A análise mostrou que fatores como salário, posse de imóvel e histórico de bom pagador aumentam o limite de crédito, enquanto inadimplência e estado civil solteiro reduzem. Indivíduos casados e bons pagadores têm limites significativamente maiores, destacando a importância de variáveis comportamentais e de estabilidade financeira na avaliação de risco.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import matplotlib as mpl
import seaborn as sns
```

```python

data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissao': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico','Engenheiro', 'Estudante','Estudante'],
    'Salario': ['4500', '8000', '5000', '10000', '12000','15000', '1200','1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000','2000', '500','250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0','1', '0','1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento','Solteiro', 'Solteiro','Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1','1', '0','0']
}


df = pd.DataFrame(data)
df['Salario']=df['Salario'].astype(float)
df['Limite_Credito']=df['Limite_Credito'].astype(float)
df['Historico_Inadimplencia']=df['Historico_Inadimplencia'].astype(int)
df['Estado_Civil'].replace('Casamento', 'Casado',inplace=True)




df=pd.concat([df,pd.get_dummies(df['Imovel_Proprio'],prefix='Imovel_Proprio')],axis=1)
df=pd.concat([df,pd.get_dummies(df['Historico_Inadimplencia'],prefix='Inadimplencia')],axis=1)
df=pd.concat([df,pd.get_dummies(df['Estado_Civil'],prefix='Estado_Civil')],axis=1)

#remoção da colunas binárias
#df.drop(columns=['Historico_Inadimplencia','Estado_Civil','Imovel_Proprio'], inplace=True)

df.rename(columns={'Imovel_Proprio_0': 'Sem Imovel', 
                   'Imovel_Proprio_1': 'Com Imovel',
                   'Inadimplencia_0': 'Bom Pagador',
                   'Inadimplencia_1': 'Inadimplencia',
                   'Estado_Civil_Casado': 'Casado',
                   'Estado_Civil_Solteiro': 'Solteiro' }, inplace=True)
df
```

### Pergunta: Quais são as variaveis mais significativas para o limite de crédito?

O limite de crédito apresenta correlação positiva com variáveis como idade (0.34), salário(0.461), posse de imóvel(0.34), histórico de bom pagador(0.50) e casados(0.64). Por outro lado, observa-se uma correlação inversa com a ausência de imóvel(-0.34), inadimplência(-0.50) e estado civil solteiro(-0.64).
* **Correlação positiva com limite de crédito** : Idade, Salário, ter imovel,Bom Pagador, Casado
* **Correlação Inversa  com limite de crédito** : Sem Imovel,Inadiplencia, Solteiro

```python
df_corr = df[['Idade',
              'Salario',
              'Sem Imovel',
              'Com Imovel',
              'Bom Pagador',
              'Inadimplencia',
              'Casado',
              'Solteiro',
              'Limite_Credito'
            ]].corr()

#Definição padrão de cor
teal_orange = LinearSegmentedColormap.from_list("TealOrange", ["#008080", "#FFFFFF", "#FF7700"])
cmap = plt.get_cmap(teal_orange)
val=df_corr['Limite_Credito'].sort_values() #ordem crescente

norm = mpl.colors.Normalize(vmin=-1, vmax=1)
scaled_values = [2 * (v - min(val)) / (max(val) - min(val)) - 1 for v in val]


plt.figure(figsize=(12, 8))
plt.bar(val.index[0:-1], val[0:-1],color=cmap(norm(scaled_values)))
plt.title('Figura1: Correlação com limite de Crédito')
plt.xticks(rotation=45)
plt.xticks(rotation=45)
plt.colorbar(plt.cm.ScalarMappable(cmap=cmap, norm=norm), label='Valor')

val
```

### 1) Qual a diferença de crédito entre os grupos?

A concessão de limite de crédito varia entre diferentes grupos.Constatamos que indivíduos casados, quando comparado com solteiros, apresentam um limite 254% maior. De forma semelhante, o limite de crédito para bons pagadores, quando comparados com  indivíduos com algum histórico de inadimplência,  apresentam limite 284% maior.

* **Solteiros X Casados** : Pessoas casadas tem um limite de crédito médio 254% maior que pessoas solteiras
* **Bom pagador x Inadimplentes  com limite de crédito** : Bons pagadores tem um limite de crédito 284% maior que pessoas com histórico de indadimplência 

```python
#Agrupamentos 
#limites de crédidos com estado civil,inadimplecina e salário. 
#Em média, qual a diferença entes esses grupos?
estado_civil= df.groupby('Estado_Civil')['Limite_Credito'].mean()

# Criando o gráfico de barras
fig, ax = plt.subplots(figsize=(10, 8))

bars = ax.bar(estado_civil.index, estado_civil, color=['#008080','#FF7700'])

for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.3,  # Ajuste do valor no topo
            f'{yval:.2f}', ha='center', va='bottom', fontsize=12)
plt.title('Figura2: Limite de crédito x Estado civil',fontsize=16)    
plt.ylabel('Média do crédito liberado',fontsize=14)
plt.xlabel('Estado civil',fontsize=14)
ax.tick_params(axis='x', labelsize=14)
valor=(estado_civil['Casado']/estado_civil['Solteiro']-1)*100;

texto=f'Pessoas casadas têm \num limite de crédito {valor:.2f}%\nmaior que solteiras'
ax.text(bar.get_x() + bar.get_width()/2-0.4,4000,texto,fontsize=16)
```

```python
inadimplencia= df.groupby('Historico_Inadimplencia')['Limite_Credito'].mean()
inadimplencia.index=['Bom pagador','Inadimplencia']
fig, ax = plt.subplots(figsize=(10, 8))

bars = ax.bar(inadimplencia.index, inadimplencia, color=['#008080','#FF7700'])

for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.3,  # Ajuste do valor no topo
            f'{yval:.2f}', ha='center', va='bottom', fontsize=12)
plt.title('Limite de crédito x Inadimplência',fontsize=16)    
plt.ylabel('Média do crédito liberado',fontsize=14)
plt.xlabel('Situação',fontsize=14)
ax.tick_params(axis='x', labelsize=14)
valor=(inadimplencia['Bom pagador']/inadimplencia['Inadimplencia']-1)*100

texto=f'Bons pagadores têm \num limite de crédito {valor:.2f}%\nmaior que Inadiplentes'
ax.text(bar.get_x() + bar.get_width()/2-0.4,3000,texto,fontsize=16)
```

### 2) Qual a relação entre salário e Limite de Crédito?

O limite de crédito apresenta uma relação crescente com o salário. No entanto, variáveis como inadimplência e estado civil (ser solteiro) impactam negativamente a concessão de um limite mais alto. Observa-se que, mesmo indivíduos solteiros com salários elevados tendem a possuir um limite de crédito inferior quando comparados a pessoas casadas ou classificadas como bons pagadores.


```python
salario= df.groupby('Salario')['Limite_Credito'].mean()
fig, ax = plt.subplots(figsize=(10, 8))

plt.plot(salario.index, salario,linewidth=2,color='#008080',label='Salário')
df['Inadimplencia'].replace(0, np.nan, inplace=True)

df['Solteiro'].replace(0, np.nan, inplace=True)


plt.scatter(df['Salario'], df['Solteiro']*1000 , color='blue', 
            marker="D", s=150, label='Solteiro')
plt.scatter(df['Salario'], df['Inadimplencia']*1000, color='#FF7700', 
         marker=".",s=350 , label='Inadimplência')

plt.grid(True, linestyle="--", alpha=0.6)
plt.xlabel('Salário',fontsize=14)
plt.ylabel('Crédito Liberado',fontsize=14)
plt.title('Figura3: Limite de crédito x Estado civil',fontsize=16)    

plt.legend()
```
